```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load datasets
gdi_df = pd.read_csv("/content/GDI_cleaned (1).csv")
income_df = pd.read_csv("world_bank_income.csv")

# Merge datasets on 'Country'
merged_df = pd.merge(gdi_df, income_df, on="Country", how="left")

# Drop missing entries if needed
merged_df = merged_df.dropna(subset=["GDI_Value", "Income_Group"])

# Boxplot: GDI distribution across income groups
plt.figure(figsize=(10, 6))
sns.boxplot(data=merged_df, x="Income_Group", y="GDI_Value", palette="Set3")
plt.title("GDI Distribution Across World Bank Income Groups")
plt.xlabel("Income Group")
plt.ylabel("GDI Value")
plt.xticks(rotation=15)
plt.tight_layout()
plt.savefig("gdi_income_group_boxplot.png")
plt.show()
```

```python
summary = merged_df.groupby("Income_Group")["GDI_Value"].describe()
print(summary)
```
